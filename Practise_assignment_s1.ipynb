```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the dataset
data_path = 'winequality-red.csv'  # Update this if the file is in a different location
wine_data = pd.read_csv(data_path)

# Function for KMeans clustering and elbow method
def kmeans_elbow_analysis(data, features, max_clusters=10):
    """
    Perform KMeans clustering and plot the elbow curve for the given features.

    Args:
    - data (pd.DataFrame): Dataset containing the features.
    - features (list): List of feature names to include in the analysis.
    - max_clusters (int): Maximum number of clusters to test.

    Returns:
    - None (plots elbow curve and prints optimal k).
    """
    # Extract relevant features
    subset = data[features]
    
    # Scale the features
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(subset)
    
    # Apply KMeans and compute WCSS for different k values
    wcss = []
    cluster_range = range(1, max_clusters + 1)
    for k in cluster_range:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(scaled_data)
        wcss.append(kmeans.inertia_)
    
    # Plot the elbow curve
    plt.figure(figsize=(8, 5))
    plt.plot(cluster_range, wcss, marker='o', linestyle='--', color='b')
    plt.title(f"Elbow Method for Features: {features}")
    plt.xlabel("Number of Clusters")
    plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
    plt.grid()
    plt.show()

# Feature groups as per the problem statement
feature_groups = [
    ["fixed_acidity", "volatile_acidity"],
    ["citric_acid", "fixed_acidity"],
    ["residual_sugar", "sulphates"],
    ["free_sulfur_dioxide", "total_sulfur_dioxide"],
    ["fixed_acidity", "citric_acid", "volatile_acidity"],
    ["density", "pH"]
]

# Perform KMeans elbow analysis for each feature group
for group in feature_groups:
    print(f"Analyzing features: {group}")
    kmeans_elbow_analysis(wine_data, group, max_clusters=10)
```
